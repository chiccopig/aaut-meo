In the following we start using a dataset (from UCI Machine Learning repository)

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

# Declare the type of prediction model and the working criteria for the model induction algorithm

```python
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:1})
```

# Split the dataset in training and test set

```python
# Generate a random permutation of the indices of examples that will be later used
# for the training and the test set
import numpy as np
np.random.seed(0)
indices = np.random.permutation(len(iris.data))

# We now decide to keep the last 10 indices for test set, the remaining for the training set
indices_training=indices[:-10]
indices_test=indices[-10:]

iris_X_train = iris.data[indices_training] # keep for training all the matrix elements with the exception of the last 10
iris_y_train = iris.target[indices_training]
iris_X_test  = iris.data[indices_test] # keep the last 10 elements for test set
iris_y_test  = iris.target[indices_test]
```

# Fit the learning model on training set

```python
# fit the model to the training data
clf = clf.fit(iris_X_train, iris_y_train)
```

# Obtain predictions

```python
# apply fitted model "clf" to the test set
predicted_y_test = clf.predict(iris_X_test)

# print the predictions (class numbers associated to classes names in target names)
print("Predictions:")
print(predicted_y_test)
print("True classes:")
print(iris_y_test)
print(iris.target_names)
```

Print the index of the test instances and the corresponding predictions

```python
# print the corresponding instances indexes and class names
for i in range(len(iris_y_test)):
    print("Instance # "+str(indices_test[i])+": ")
    print("Predicted: "+iris.target_names[predicted_y_test[i]]+"\t True: "+iris.target_names[iris_y_test[i]]+"\n")
```

# Look at the specific examples

```python
for i in range(len(iris_y_test)):
    print("Instance # "+str(indices_test)+": ")
    s=""
    for j in range(len(iris.feature_names)):
        s=s+iris.feature_names[j]+"="+str(iris_X_test[i][j])
        if (j<len(iris.feature_names)-1): s=s+", "
    print(s)
    print("Predicted: "+iris.target_names[predicted_y_test[i]]+"\t True: "+iris.target_names[iris_y_test[i]]+"\n")
```

# Obtain model performance results

```python
# print some metrics results
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
acc_score = accuracy_score(iris_y_test, predicted_y_test)
print("Accuracy score: "+ str(acc_score))
f1=f1_score(iris_y_test, predicted_y_test, average='macro')
print("F1 score: "+str(f1))
```

# Use Cross Validation

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score # will be used to separate training and test
iris = load_iris()
clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:1})
clf = clf.fit(iris.data, iris.target)
scores = cross_val_score(clf, iris.data, iris.target, cv=5) # score will be the accuracy
print(scores)
```

```python
# computes F1- score
f1_scores = cross_val_score(clf, iris.data, iris.target, cv=5, scoring='f1_macro')
print(f1_scores)
```

```python
#1a artificial inflation

inf_X_virginica = np.repeat(iris_X_train[iris_y_train==2],10,axis=0) # inflation su virginica (dati)
iris_X = np.concatenate((iris_X_train, inf_X_virginica), axis=0) # inflation training set

inf_Y_virginica = np.repeat(iris_y_train[iris_y_train==2],10,axis=0) # inflation su virginica (target)
iris_y = np.concatenate((iris_y_train, inf_Y_virginica), axis=0)

# rirandomizzazione indici per X e y
np.random.seed(0)
r_indices = np.random.permutation(len(iris_X))
iris_X = iris_X[r_indices]
iris_y = iris_y[r_indices]

# addestramento del modello DT con entropia come criterio di split
inf_clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:1})
inf_clf = inf_clf.fit(iris_X, iris_y)

# predizione sul test set
inf_prediction = inf_clf.predict(iris_X_test)
print("Predictions:")
print(inf_prediction)
print("True classes:")
print(iris_y_test)

# calcolo accuracy e f1-score
acc_score = accuracy_score(iris_y_test, inf_prediction)
print("Accuracy score: "+ str(acc_score))
f1=f1_score(iris_y_test, inf_prediction, average='macro')
print("F1 score: "+str(f1))

# visualizzazione dell'albero
dot_data = tree.export_graphviz(inf_clf, out_file=None,
                         feature_names=iris.feature_names,
                         class_names=iris.target_names,
                         filled=True, rounded=True,
                         special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
#1b modifying the weights

# modifica dei pesi della classe virginica
w_clf = tree.DecisionTreeClassifier(criterion="entropy",random_state=300,min_samples_leaf=5,class_weight={0:1,1:1,2:10})
w_clf = w_clf.fit(iris_X_train, iris_y_train)

# predizione sul test set
w_prediction = w_clf.predict(iris_X_test)
print("Predictions:")
print(w_prediction)
print("True classes:")
print(iris_y_test)

# calcolo accuracy e f1 score
acc_score = accuracy_score(iris_y_test, w_prediction)
print("Accuracy score: "+ str(acc_score))
f1=f1_score(iris_y_test, w_prediction, average='macro')
print("F1 score: "+str(f1))

# visualizzazione dell'albero
dot_data = tree.export_graphviz(w_clf, out_file=None,
                         feature_names=iris.feature_names,
                         class_names=iris.target_names,
                         filled=True, rounded=True,
                         special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
#2 GridSearchCV

from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report

# definizione griglia di ricerca (4x4x3x4=192 combinazioni di parametri)
param_grid = {
    'max_depth' : [3,5,10,None], # limita profondità massima
    'min_samples_leaf' : [1,2,5,10], # n minimo di campioni in una foglia
    'min_impurity_decrease' : [0.0,0.01,0.1], # soglia minima per decremento impurezza
    'max_leaf_nodes' : [None,10,20,50] # n massimo di foglie
}

# creazione albero decisionale
tun_clf = tree.DecisionTreeClassifier(random_state=42)

# grid search con cross-validation
grid_search = GridSearchCV(
    estimator=tun_clf,
    param_grid=param_grid,
    scoring='accuracy',
    cv=5,
    verbose=1, # stampa dei dettagli
    n_jobs=-1  # usa tutti i core disponibili per velocizzare il processo
)

# addestramento del modello ottimizzato (tuning iper-parametri)
grid_search.fit(iris_X_train, iris_y_train)

# stampa dei risultati della cross-validation
print(grid_search.best_params_)
print("Cross-val: ", grid_search.best_score_)

# test del modello ottimizzato
best_model = grid_search.best_estimator_
tun_pred = best_model.predict(iris_X_test)

print("\nPerformance sul test set:")
print("Accuracy", accuracy_score(iris_y_test, tun_pred))
print("Classification report:\n", classification_report(iris_y_test, tun_pred))

# visualizzazione albero decisionale
dot_data = tree.export_graphviz(best_model, out_file=None,
                         feature_names=iris.feature_names,
                         class_names=iris.target_names,
                         filled=True, rounded=True,
                         special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
#3 matrice di confusione

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# funzione generazione matrice di confusione
def plt_confusion_matrix(y_true, y_pred, labels, title):
  cm = confusion_matrix(y_true, y_pred)
  disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
  disp.plot(cmap=plt.cm.Blues)
  plt.title(title)
  plt.show()

plt_confusion_matrix(iris_y_test, predicted_y_test, iris.target_names, "Matrice di Confusione - Original")
plt_confusion_matrix(iris_y_test, inf_prediction, iris.target_names, "Matrice di Confusione - Inflated")
plt_confusion_matrix(iris_y_test, w_prediction, iris.target_names, "Matrice di Confusione - Weighted")
plt_confusion_matrix(iris_y_test, tun_pred, iris.target_names, "Matrice di Confusione - Tuned")
```

```python
#4 ROC Plot

from sklearn.metrics import roc_curve, auc

# definizione dei modelli
clfs = [inf_clf, w_clf, best_model]

# creazione figure
fig, axs = plt.subplots(len(clfs), 3, figsize=(15,25))

for i in range(len(clfs)) : # i modelli
  for j in range(3): # j classi
    predictions = clfs[i].predict(iris_X_test)
    fpr, tpr, _ = roc_curve(np.where(iris_y_test==j, 1, -1), np.where(predictions==j, 1, -1)) # classe j vs rest
    axs[i,j].plot(fpr, tpr)
    axs[i,j].set_title(f'Model : {i} | ROC curve {iris.target_names[j]} VS others')

for ax in axs.flat:
  ax.set(xlabel='fpr', ylabel='tpr')

for ax in axs.flat:
    ax.label_outer()

fig.show()
```
