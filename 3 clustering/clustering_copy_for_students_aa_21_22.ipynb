## In the following program we cluster the dataset n.1 with K-means.

```python
import csv
import sys
from os.path import join

import numpy as np

def load_data(file_path, file_name):
   with open(join(file_path, file_name)) as csv_file:
       data_file = csv.reader(csv_file,delimiter=',')
       temp1 = next(data_file)
       n_samples = int(temp1[0])
       n_features = int(temp1[1])
       temp2 = next(data_file)
       feature_names = np.array(temp2[:n_features])

       data_list = [iter for iter in data_file]

       data = np.asarray(data_list, dtype=np.float64)

   return(data,feature_names,n_samples,n_features)

# The main program reads the input file containing the dataset
# file_path is the file path where the file with the data to be read are located
# we assume the file contains an example per line
# each example is a list of real values separated by a comma (csv format)
# The first line of the file contains the heading with:
# N_samples,n_features,
# The second line contains the feature names separated by a comma

#file_path="~/meo/Documents/Didattica/Laboratorio-15-16-Jupyter/"
file_path="./sample_data/"
# all the three datasets contain data points on (x,y)
file_name1="3-clusters.csv"

data1,feature_names1,n_samples1,n_features1 = load_data(file_path, file_name1)

from sklearn.cluster import KMeans

np.random.seed(5)

k=3
kmeans1 = KMeans(n_clusters=k, random_state=0).fit(data1)

i=0
for i in range(n_samples1):
    print("Example n."+str(i)+"=("+str(data1[i,0])+","+str(data1[i,1])+")")
    print("in cluster n."+str(kmeans1.labels_[i]))
```

In the following cells I propose you to write a program that computes with a statistical measure of your choice a quantitative evaluation of clusters in the three datasets.

**Note:**
It is advisable to execute K-means a certain number of times (let us try 10 times) and then select the clustering solution that gives the best value of the evaluation measure.

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# funzione per eseguire K-means 10 volte e selezionare il miglior risultato
def evaluate_clusters(data, n_clusters, n_init=10):
    best_score = -1
    best_kmeans = None

    for i in range(n_init):
        # esecuzione K-means
        kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=i)
        labels = kmeans.fit_predict(data)

        # calcolo silhouette score
        score = silhouette_score(data, labels)
        print(f"Iterazione {i+1}: Silhouette Score = {score}")

        # aggiornamento del miglior risultato
        if score > best_score:
            best_score = score
            best_kmeans = kmeans

    # memorizzazione del modello k-means con il punteggio migliore
    return best_kmeans, best_score
```

```python
# valutazione dataset 1
n_clusters = 3

print(f"\nValutazione per Dataset 1")
best_kmeans, best_score = evaluate_clusters(data1, n_clusters)

print(f"Miglior Silhouette Score per Dataset 1: {best_score}")
```

```python
# valutazione dataset 2
n_clusters = 3

print(f"\nValutazione per Dataset 2")
best_kmeans, best_score = evaluate_clusters(data2, n_clusters)

print(f"Miglior Silhouette Score per Dataset 2: {best_score}")
```

```python
# valutazione dataset 3 campionato

# campionamento
n_samples = int(len(data3) * 0.1)
indices = np.random.choice(len(data3), n_samples, replace=False)

n_clusters = 7

print(f"\nValutazione per Dataset 3 (campionato)")
best_kmeans, best_score = evaluate_clusters(data3[indices], n_clusters)

print(f"Miglior Silhouette Score per Dataset 3: {best_score}")
```

As already done with classification by k-nn, plot in the cell below the quantitative measure of your choice (used above) with respect to an increasing value of k (the number of clusters) so that the best value of k can be selected.

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import numpy as np

# funzione per calcolare silhouette score per valori crescenti di k
def silhouette_analysis(data, k_range):
    scores = []
    best_k = None
    best_score = -1

    for k in k_range:
        # esecuzione K-means
        kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
        labels = kmeans.fit_predict(data)

        # calcolo silhouette score
        score = silhouette_score(data, labels)
        scores.append(score)

        # aggiornamento miglior k (se necessario)
        if score > best_score:
            best_score = score
            best_k = k

    return scores, best_k, best_score

# funzione per analizzare un dataset e opzionalmente plottare i risultati
def analyze_and_plot(data, dataset_name, k_values, plot=True):
    print(f"\nAnalisi del Silhouette Score per {dataset_name}")
    scores, best_k, best_score = silhouette_analysis(data, k_values)

    print(f"\nMiglior numero di cluster (k): {best_k}")
    print(f"Silhouette Score corrispondente: {best_score}")

    if plot:
        plt.figure(figsize=(10, 6))
        plt.plot(k_values, scores, marker='o', label="Silhouette Score")
        plt.axvline(best_k, color='r', linestyle='--', label=f"Miglior k = {best_k}")
        plt.title(f"Silhouette Score vs Numero di cluster (k) - {dataset_name}")
        plt.xlabel("Numero di cluster (k)")
        plt.ylabel("Silhouette Score")
        plt.grid(True)
        plt.xticks(k_values)
        plt.legend()
        plt.show()

# definizione del range di k
k_values = range(2, 11)  # prova valori di k da 2 a 10

# analisi per i tre dataset
datasets = {
    "Dataset 1": data1,
    "Dataset 2": data2,
    "Dataset 3": data3[indices]
}

for name, dataset in datasets.items():
    analyze_and_plot(dataset, name, k_values)
```

```python
# DBSCAN per dataset 2

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# applicazione DBSCAN (parametri a scelta)
dbscan = DBSCAN(eps=3, min_samples=10)
labels = dbscan.fit_predict(data2)

# conta del numero di cluster trovati e dei punti rumore (esclusi)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Numero di cluster trovati: {n_clusters}")
print(f"Numero di noise points: {n_noise}")

# plot dei cluster
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = [plt.cm.jet(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, color in zip(unique_labels, colors):
    if label == -1:
        color = (0.8, 0.8, 0.8, 1)  # grigio chiaro per il rumore

    # selezione dei punti appartenenti al cluster
    class_mask = labels == label
    plt.scatter(data2[class_mask, 0], data2[class_mask, 1], c=[color], s=10, label=f"Cluster {label}")

plt.title(f"DBSCAN Clustering (eps={dbscan.eps}, minPts={dbscan.min_samples})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```

In the following cell I propose you to:

1.   Set *Minpts* to a number (say **10**).
2.   Compute the **reachability distance** of the **10-th nearest neighbour ** for each data-point.
3. Sort the set of reachability distances you obtained in an increasing way.
4. Plot the sorted reachability distances
5. Find the **elbow** of the diagram => it gives the eps value combined with Minpts=10.
6. Try this combined pair of parameters on the dataset you chose, with DBSCAN.

```python
# reachability distance x dataset 2

from sklearn.neighbors import NearestNeighbors

# 1 set minpts a 10
min_pts = 10

# 2 calcolo della reachability distance per il 10imo vicino più vicino per ciascun punto
neighbors = NearestNeighbors(n_neighbors=min_pts)
neighbors = neighbors.fit(data2)
distances = neighbors.kneighbors(data2)[0]

# 3 ordinamento delle reachability distances
distances = np.sort(distances[:, min_pts-1], axis=0)

# 5 elbow
elbow = np.diff(distances) # guarda le distanze tra punti consecutivi
max_elbow = np.argmax(elbow) # prende la massima curvatura
print(f'Il punto di elbow si trova a reachability distance = {distances[max_elbow]}')

#4 plot
plt.plot(distances)
plt.title('Reachability distances')
plt.xlabel('Punti ordinati secondo la reachability distance')
plt.ylabel('Reachability distance')
plt.scatter(max_elbow, distances[max_elbow], color='red')
plt.show()
```

```python
# DBSCAN per dataset 2 con eps determinato dal punto di elbow

import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler

# applicazione DBSCAN
dbscan = DBSCAN(eps=distances[max_elbow], min_samples=10)
labels = dbscan.fit_predict(data2)

# conta del numero di cluster trovati e dei punti rumore (esclusi)
n_clusters = len(set(labels)) - (1 if -1 in labels else 0)
n_noise = list(labels).count(-1)

print(f"Numero di cluster trovati: {n_clusters}")
print(f"Numero di noise points: {n_noise}")

# plot dei cluster
plt.figure(figsize=(10, 6))
unique_labels = set(labels)
colors = [plt.cm.jet(each) for each in np.linspace(0, 1, len(unique_labels))]

for label, color in zip(unique_labels, colors):
    if label == -1:
        color = (0.8, 0.8, 0.8, 1)  # grigio chiaro per il rumore

    # seleziona dei punti appartenenti al cluster
    class_mask = labels == label
    plt.scatter(data2[class_mask, 0], data2[class_mask, 1], c=[color], s=10, label=f"Cluster {label}")

plt.title(f"DBSCAN Clustering (eps={dbscan.eps}, minPts={dbscan.min_samples})")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.legend()
plt.show()
```
